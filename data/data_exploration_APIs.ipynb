```python
stations = {'St. Petersburg':"8726520"}
```

```python
def read_noaa_water_temp(station_num, start_date, end_date, interval_length = 1):
    url = "https://api.tidesandcurrents.noaa.gov/api/prod/datagetter?product=water_temperature&application=NOS.COOPS.TAC.PHYSOCEAN&begin_date="+start_date+"&end_date="+end_date+"&station="+station_num+"&time_zone=GMT&units=english&interval="+str(interval_length)+"&format=json"
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Error: Received status code {response.status_code}")
        return url
    df = pd.DataFrame(response.json()['data'])
    df.rename(columns = {"time":"t","temp_F":"v","conductivitity":"f"})
    return df
```

```python
read_noaa_water_temp(stations['St. Petersburg'], '20240620','20240621', 1000)
```
